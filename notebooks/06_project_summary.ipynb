# Fortnite Player Skill Modeling

### 1. Project Overview

This project investigates whether player behavior in Fortnite can be
mathematically modeled and used to derive an interpretable Skill Index.

The analysis combines:

1. Behavioral feature engineering
2. Dimensionality reduction
3. Player archetype discovery
4. Skill index construction
5. Cross-mode generalization analysis

The goal is not only predictive performance but also
interpretability of gameplay structure.

### 2. Dataset Description

The dataset contains performance statistics for Fortnite players across
multiple game modes.

Available modes include:

- Solo
- Duos
- Trios
- Squads
- Limited Time Modes (LTM)

For each mode, statistics include:

- Score
- Win count
- Kill/death ratio
- Win rate
- Matches played
- Total eliminations
- Time played

After stability filtering (Solo matches ≥ 50), the analysis includes
1433 players.

### 3. Behavioral Feature Engineering

Raw statistics are transformed into behavior-oriented metrics:

1. Solo_kpm — kills per match
2. Solo_mpm — minutes played per match
3. Solo_wr — win rate

These transformations remove the influence of raw play volume and
allow analysis of gameplay tendencies rather than activity level.

### 4. Latent Structure of Player Behavior

Principal Component Analysis reveals that player behavior is largely
two-dimensional.

PC1 (67% variance explained)

Represents combat dominance:

- high KD
- high kills per match

PC2

Represents survival tendency:

- longer match duration
- slower gameplay style

This indicates that Fortnite Solo gameplay can be interpreted
as a combination of combat efficiency and survival behavior.

### 5. Behavioral Archetypes

Clustering in PCA space identifies three distinct player archetypes:

1. Low-impact players
   - low combat efficiency
   - low survival performance

2. Combat-dominant players
   - high KD
   - highest win rate

3. Survival-oriented players
   - longer match duration
   - moderate success rates

These archetypes demonstrate that player behavior is not purely
continuous but exhibits meaningful segmentation.

### 6. Skill Index Construction

The first principal component (PC1) is used as the Skill Index.

SkillIndex = PC1

This index captures the dominant behavioral axis associated with
combat efficiency.

Regression analysis shows that behavioral variables explain
approximately 87% of the variance in Solo win rate.

### 7. Skill Transfer Across Game Modes

We evaluate whether Solo-derived skill transfers to team-based modes.

Regression results:

Duos
R² ≈ 0.36

Trios
R² ≈ 0.14

Squads
R² ≈ 0.58

These results suggest that individual combat efficiency remains an
important component of team-mode performance, but its influence varies
depending on team size and coordination dynamics.

### 8. Key Insights

1. Player behavior in Fortnite has a clear latent structure.
2. Combat efficiency is the dominant driver of Solo success.
3. Player populations exhibit identifiable gameplay archetypes.
4. A simple PCA-derived Skill Index effectively summarizes player ability.
5. Individual skill partially transfers to team-based gameplay but is
   moderated by team dynamics.


Overall, this project demonstrates how linear algebra and statistical
modeling can reveal interpretable structure in complex behavioral data.
